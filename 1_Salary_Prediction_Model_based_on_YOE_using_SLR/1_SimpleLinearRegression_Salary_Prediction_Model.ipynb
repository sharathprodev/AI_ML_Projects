```python
#importing library to handle tabular datasets
import pandas as pd 
```

```python
dataset=pd.read_csv("salary_data.csv")
```

```python
dataset
```

```python
#setting up independent variable
independent_variable = dataset[["YearsExperience"]]
independent_variable
```

```python
#setting up dependent variable
dependent_variable = dataset[["Salary"]]
dependent_variable
```

```python
#importing train_test_split from sklearn library to split dataset eg:for training ku 80% and testing ku 20%
from sklearn.model_selection import train_test_split
# sklearn library contains a lot of efficient tools for machine learning and statistical modeling including classification,
#regression, clustering and dimensionality reduction
```

```python
#train_test_split is providing us four outputs and we are storing them in below variables
# we are setting test_size=0.30 means 70% out of our dataset for traing and 30% will be used to testing (to calculate the evaluation metrics)
x_train,x_test,y_train,y_test=train_test_split(independent_variable, dependent_variable, test_size=0.30, random_state=0)
```

```python
x_train #YearsExperience Dataset Split for Training the model
```

```python
x_test #YearsExperience Dataset Split for Testing the model
```

```python
y_train #Salary Dataset Split for Training the model
```

```python
y_test #Salary Dataset Split for Testing the model
```

```python
from sklearn.linear_model import LinearRegression
regressor_variable=LinearRegression() #Assigning LinearRegression methods to variable to refernce with regressor_variable
regressor_variable.fit(x_train,y_train) #Passing Training Set's fit(input dataset,output dataset) in fit method
#this is the learning phase of this model
#this line will be common in most of ML models
```

```python
#assigning output from LinearRegression's 
weight=regressor_variable.coef_
weight
```

```python
bais=regressor_variable.intercept_
bais
```

```python
#using predict
y_predict=regressor_variable.predict(x_test)
# Use the .predict() method on new data to generate predictions. 
#The input to this method should be a 2D array-like structure where each row represents a sample 
#and each column represents a feature.
y_predict
```

```python
#Testing the Model (Evalution Metrics)
from sklearn.metrics import r2_score
r_score= r2_score(y_test,y_predict)   #y_test is test data split for testing(which is actual data available in dataset) 
#vs y_predit (AI Model) predited Data
r_score
```

```python
#r_score value ranges from 0 to 1 and a good model should be near to 1
#once the model achieves the r_score value near to we can start to save the model
```

```python
#Saving the model using Pickle
import pickle
filename="finalized_model_SLR_Sal_Pred_Model.sav"
```

```python
pickle.dump(regressor_variable,open(filename,'wb'))  #wb refers to write binary
```

```python
#Now if you check the folder, model will be saved
#then lets valdiate whether saved model works properly
```

```python
loaded_model=pickle.load(open(filename,'rb'))
```

```python
result=loaded_model.predict([[15]]) #Testing what should be the salary for 15 Years of experiance
result
```
