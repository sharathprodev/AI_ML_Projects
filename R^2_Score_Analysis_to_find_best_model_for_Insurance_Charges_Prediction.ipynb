```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
input_var=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
output_var=dataset['charges']
```

```python
input_var
```

```python
output_var
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_var,output_var,test_size=1/3, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#1# MLR_Model_Creation
from sklearn.linear_model import LinearRegression
MLR_regressor_variable=LinearRegression()
MLR_regressor_variable.fit(x_train,y_train) 
```

```python
#we got weight and bias
weight=MLR_regressor_variable.coef_
bais=MLR_regressor_variable.intercept_
#weight = array([  260.1423112 ,   315.22441969,   545.72248029,   -71.76915955,       23252.13608407])
#bais = -12013.76012735222
```

```python
weight
```

```python
bais
```

```python
#Predition test
y_predict=MLR_regressor_variable.predict(x_test)
y_predict
```

```python
#Evaluate Predition metrics
from sklearn.metrics import r2_score
r_score= r2_score(y_test,y_predict)   
r_score
#MLR_r^2_Score=0.7865108093853883
```

```python
#we are going to standardization as the model got bad r2_value
#so after standardization we shall check whether r2_value improved
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
#fit_transform will train model with standardization
x_test=sc.transform(x_test)
```

```python
#1# MLR_Model_Creation
from sklearn.linear_model import LinearRegression
MLR_regressor_variable=LinearRegression()
MLR_regressor_variable.fit(x_train,y_train) 
```

```python
#Predition test
y_predict=MLR_regressor_variable.predict(x_test)
y_predict
```

```python
#Evaluate Predition metrics
from sklearn.metrics import r2_score
r_score= r2_score(y_test,y_predict)   
r_score
#After_Standarization_MLR_r^2_Score=0.7865108093853883 remains same
```
