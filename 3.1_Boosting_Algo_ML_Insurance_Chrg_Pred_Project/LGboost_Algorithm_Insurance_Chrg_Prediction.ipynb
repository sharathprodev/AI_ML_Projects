```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
input_var=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
output_var=dataset['charges']
```

```python
input_var
```

```python
output_var
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_var,output_var,test_size=1/3, random_state=0)
```

```python
#The difference in how data is handled and how the model is initialized between LightGBM and XGBoost arises 
#from the design and flexibility of each library.

#LightGBM
#Data Conversion: LightGBM requires data to be in a specific format (lgb.Dataset) for its training function (lgb.train). 
#This conversion is necessary to optimize memory usage and training speed, as LightGBM uses a histogram-based approach.
#Parameter Setting: Parameters in LightGBM need to be explicitly set before training because 
#LightGBM has a more complex set of hyperparameters that control various aspects of the boosting process.
#Other Models
#No Explicit Data Conversion: Previous models can handle data directly in NumPy array, pandas DataFrame, or other common formats. 
#It does not require a special dataset format for basic usage, making it simpler to use in many scenarios.
#Fit Method: XGBoost provides a fit method in the XGBRegressor class that handles both training and evaluation internally, 
#making it similar to scikit-learn's API. This simplifies the process for standard regression tasks.

#Step for LightGBM
#Data Preparation - Convert data to lgb.Dataset format
#Parameter Setting - Explicitly set parameters in a dictionary
#Training - Use lgb.train function with datasets and parameters
```
