```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
input_var=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
output_var=dataset['charges']
```

```python
input_var
```

```python
output_var
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_var,output_var,test_size=1/3, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
import xgboost as xgb

# Initialize XGBoost regressor
regressor = xgb.XGBRegressor(objective ='reg:squarederror', n_estimators = 100, random_state = 0)

# Fit the regressor to the training data
regressor.fit(x_train, y_train)

# Make predictions on the test data
XGB_y_pred = regressor.predict(x_test)

# Calculate R^2 score
from sklearn.metrics import r2_score
XGB_r_score = r2_score(y_test, XGB_y_pred)

print("XGBoost R^2 Score:", XGB_r_score)
```
