```python
#we are going to standardization as the model got bad r2_value (0.78651)
#so after standardization we shall check whether r2_value improved
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
input_var=sc.fit_transform(input_var)
#fit_transform will train model with standardization
#x_test=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
model = LinearRegression()
parameters = {'fit_intercept':[True,False], 'normalize':[True,False], 'copy_X':[True, False]}
grid = GridSearchCV(model,parameters, cv=None)
grid.fit(input_var, output_var)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
reg = Ridge()
parameters = {
    "alpha": [1, 10, 100, 290, 500],
    "fit_intercept": [True, False],
    "solver": ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
}
grid = GridSearchCV(estimator=reg, param_grid=parameters, cv=2, n_jobs=-1)
grid.fit(input_var, output_var)
best_model = grid.best_estimator_
print("Best model parameters:", grid.best_params_)
print("Test score:", best_model.score(input_var, output_var))
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```
